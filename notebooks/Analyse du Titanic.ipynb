```python
# importation des bibliothèques nécéssaires à notre analyse
import pandas as pd
import numpy as np
```

```python
# lecture du jeu de données sous forme de dataframe
df = pd.read_csv('/content/titanic.csv')
```

```python
#Aperçu des données
df.head()
```

```python
# Vérifions le nombre de lignes et de colonnes que contient notre dataframe
df.shape
```

```python
# Un aperçu des types des colonnes et des valeurs nulles
df.info()
```

```python
#Répartition des valeurs nulles
df.isnull().sum()
```

```python
# statistiques descriptives des colonnes numériques
df.describe()
```

```python
# statistiques descriptives de toutes les colonnes
df.describe(include = 'all')
```

```python
#Vérifions si notre dataset contient des doublons
df['PassengerId'].duplicated().sum()
```

```python
#Quel est le nombre de personnes ayant survécu?
survived_count = df['Survived'].sum()
print('Le nombre de personnes ayant survécu est', survived_count)
```

```python
#Quelle est la répartition du genre des passagers en nombre?

df['Sex'].value_counts()
```

```python
#Quelle est la répartition du genre des passagers en pourcentage?
df['Sex'].value_counts(normalize = True)
```

```python
#Age moyen des passagers
df['Age'].mean()
```

```python
#Age médian des passagers
df['Age'].median()
```

```python
# Prix moyen du ticket
df['Fare'].mean()
```

```python
#Prix médian du ticket
df['Fare'].median()
```

```python
#Les passagers ayant payé les tickets aux prix les plus élevés?
df[df['Fare'] == df['Fare'].max()]
```

```python
#La répartition des survivants par genre
#cat var - gender/sex
#num var - survival
#agg func - mean

df.groupby('Sex')['Survived'].mean()
```

```python
#La répartition des survivants par classe de ticket
df.groupby('Pclass')['Survived'].mean()
```

```python
#La répartition des survivants par classe et par genre
df.groupby(['Pclass','Sex'])['Survived'].mean()
```

```python
#Le profil des familles à bord
# ('SibSp' is siblings/spouses and 'Parch' is parents/children)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  # +1 for the passenger themselves
df['FamilySize'].describe()
```

```python
#Comment sont répartis les passagers selon le lieu d'embarcation?
df['Embarked'].value_counts()
```

```python
# Le nombre de survivants ayant une cabine
df[df['Survived'] == 1]['Cabin'].notna().sum()
```

```python
#Le prix moyen d'un ticket
df['Fare'].mean()
```

```python
#Le prix moyen d'un ticket par classe
df.groupby('Pclass')['Fare'].mean()
```

```python
#Le nombre d'enfants ayant survécu (passagers de moins de 18 ans)
# (Assuming 'Age' < 18 defines a child)
df[(df['Survived'] == 1) & (df['Age'] < 18)].shape[0]
```

```python
#Qui sont ceux qui avaient plus de 70 ans parmi les passagers à bord?
df[df['Age'] > 70].shape[0]
```
